```python
import anndata
import pandas as pd
```

```python
human_dict = {"path": "original_or_past_data/human.h5ad", "type": "cell_type", "val": "mast cell", "writes": "human_mast_cells.h5ad"}
ys_dict = {"path": "original_or_past_data/ys_portal_object_20221208.h5ad", "type": "LVL2", "val": "MAST_CELL", "writes": "ys_mast_cells.h5ad"}
ys_pt_dict = {"path": "original_or_past_data/ys_portal_object_20221208.h5ad", "type": "LVL3", "val": ['MAST_CELL', 'MEMP', "PRIM_HSPC_1", "PRIM_HSPC_2", "DEF_HSPC_2", "DEF_HSPC_1", "HSPC_1", 
           "HSPC_2",'CMP',"EO_BASO_MAST_PRECURSOR"], "writes": "ys_pseudotime.h5ad"}
sum_dict = {"human":human_dict, "ys":ys_dict, "ys_pt": ys_pt_dict}
```

```python
# CHANGE: select between "human", "ys", and "ys_pt" for subsetting
# no need to change everything else
dict_key = "ys_pt"
dict_to_use = sum_dict[dict_key]
```

```python
adata_path = dict_to_use["path"]
adata = anndata.read_h5ad(adata_path)
print(adata)
```
